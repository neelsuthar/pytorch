```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
# tensors
X_train_tensor = torch.tensor(X_train, dtype=torch.float64)
X_test_tensor = torch.tensor(X_test, dtype=torch.float64)

y_train_tensor = torch.tensor(y_train, dtype=torch.float64).view(-1, 1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float64).view(-1, 1)
```

```python
X_train_tensor.shape
```

```python

#define the model for our NN
```

```python
class my_simple_nn():
    def __init__(self, X):
        self.weights = torch.rand(
            X.shape[1], 1, dtype=torch.float64, requires_grad=True
        )
        self.bias = torch.zeros(
            1, dtype=torch.float64, requires_grad=True
        )

    def forward(self, X):
        z = torch.matmul(X, self.weights) + self.bias
        y_pred = torch.sigmoid(z)
        return y_pred

    def loss_function(self, y_pred, y):
        loss = torch.nn.functional.binary_cross_entropy(y_pred, y)
        return loss


```

```python
learning_rate=0.1
epochs=250
```

```python
#create my model
model=my_simple_nn(X_train_tensor)

for epoch in range(epochs):
  #fwd
  y_pred = model.forward(X_train_tensor)
  # print(y_pred)

  #loss
  loss = model.loss_function(y_pred, y_train_tensor)



  #backward
  loss.backward()

  #updation
  with torch.no_grad():
    model.weights-=learning_rate*model.weights.grad
    model.bias-=learning_rate*model.bias.grad

  #for next epoch new gradient
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  print(f"this the loss for epoch:{epoch+1} : {loss.item()}")
```

```python
#model evaluate
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred=(y_pred>0.6).float()
  accuracy=y_pred.eq(y_test_tensor).sum()/float(y_test_tensor.shape[0])
  print(f"accuracy:{accuracy.item()*100}")
```

```python
import torch
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv")
df.head()
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
df.head()
df.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
encoder=LabelEncoder()
y_train=encoder.fit_transform(y_train)
y_test=encoder.transform(y_test)
```

```python
X_train_tensor = torch.tensor(X_train, dtype=torch.float64)
X_test_tensor = torch.tensor(X_test, dtype=torch.float64)
X_test_tensor.shape
```

```python
y_train_tensor = torch.tensor(y_train, dtype=torch.float64).view(-1,1)
y_test_tensor = torch.tensor(y_test, dtype=torch.float64).view(-1,1)
```

```python
class my_simple_nn():

  def __init__(self, X):
    self.weights = torch.rand(X.shape[1],1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    loss = torch.nn.functional.binary_cross_entropy(y_pred, y);
    return loss
```

```python
learning_rate = 0.1
epochs = 250
```

```python
model=my_simple_nn(X_train_tensor)

for epoch in range(epochs):
  #fwd
  y_pred = model.forward(X_train_tensor)
  #loss
  loss=model.loss_function(y_pred, y_train_tensor)
  #bck
  loss.backward()
  #updation
  with torch.no_grad():
    model.weights-=learning_rate*model.weights.grad
    model.bias-=learning_rate*model.bias.grad

    model.weights.grad.zero_()
    model.bias.grad.zero_()

  print(f"this the loss for epoch:{epoch+1} : {loss.item()}")
```

```python
with torch.no_grad():
    y_pred = model.forward(X_test_tensor)
    y_pred = (y_pred > 0.5).float()
    accuracy = y_pred.eq(y_test_tensor).sum() / y_test_tensor.shape[0]
    print(f"Accuracy: {accuracy.item() * 100:.2f}%")
```
